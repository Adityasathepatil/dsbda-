```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('D:\\third year\\Sem 6\\dsbda\\practical\\datasets\\heart.csv')
```

```python
df['Sex']=df['Sex'].replace({1:"Male",0:"Female"})
df
```

```python
df=df.drop('Unnamed: 0',axis=1) #dropping unnamed as its not useful
```

```python
# a)Hitsgram
df.plot.hist()
```

```python
plt.hist(df["Age"])
```

```python
# b) Bar Chart
sns.barplot(x='ChestPain', y= "Age", data=df, ci=None, hue = "Sex") #ci=None for hiding top lines
```

```python
sns.barplot(x='ChestPain', y= "Age", data=df, ci=None, color = "Yellow" , edgecolor = "Red" ,lw = 2) #lw for edge color
```

```python
# c) Pie Chart
f,ax=plt.subplots(1,2,figsize=(10,6))
df["AHD"].value_counts().plot.pie(ax = ax[0],autopct='%1.1f%%')
```

```python
# d) Box Plot
sns.boxplot(x='AHD', y='Oldpeak', data=df)
```

```python
fig=plt.figure(figsize=(20,10))
sns.boxplot(data = df,linewidth = 2.5)
plt.show()
```

```python
# e) heatmap
corr = df.corr()
corr #Pandas dataframe.corr() is used to find the pairwise correlation of all columns in the dataframe. Any na values are 
#automatically excluded. For any non-numeric data type columns in the dataframe it is ignored.
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(corr,annot= True)
```
